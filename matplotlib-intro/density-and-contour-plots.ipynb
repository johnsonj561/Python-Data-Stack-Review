### Density and Contour Plots

Contours and color-coded regions can be used to display 3D data in 2D plots.

plt.contour -> contour plots

plt.contourf -> filled contour plots

plt.imshow -> showing images

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

#### Visualizing a 3D Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 40)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

When only a single color is used, negative values are represented with dashed lines and positive values with solid lines.

Alternatively, lines can be color-coded with the cmap argument.

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
# we can explore all colors!

plt.cm.*?
```

```python
# we can reduce the number of distracting lines with a filled contour plot

plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```

```python
# the color steps are discrete, so it looks blotchy!

# another option is to use plt.imshow()
# plt.imshow() will interpret two-dimensional grid of data as an image

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

```python
# we can also combine contour plots and image plots

plt.figure(figsize=(10, 10))
contours = plt.contour(X, Y, Z, 3, colors='black')

# add labels to the contours
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```
