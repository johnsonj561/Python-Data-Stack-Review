### MatplotLib Visualization

Released in 2003.

Matplotlib is multi-platform package for data visualization built on NumPy arrays.

Was adopted as the plotting package of choice by the Space Telescope Science Institude --> The guys who made the Hubble!

It works very well with many backends and operating systems.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

#### MATLAB-style Interface

Since Matplotlib was originally written as a Python alternative for Matlab users, a lot fo the syntax is the same.

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
plt.gcf() # gets current figure
```

```python
plt.gca() # gets current axes
```

#### Object-Oriented Interface

Able to handle more complicated scenarios.

Plottings functions are methods of explicit Figure and Axes objects.

```python
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
